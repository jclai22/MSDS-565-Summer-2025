## 2. PIL

```python
from PIL import Image

# PIL's Image.open opens the image. It can be directly displayed by matplotlib.
img_pil = Image.open(image_path)
print(img_pil)
```

```python
plt.figure()
plt.imshow(img_pil)
plt.axis('off')
plt.show()
```

## 3. cv2

**cv2 loads images in BGR format and must be reversed.**

```python
import cv2

# cv2.imread loads the image in BGR format. Convert it to RGB for proper display.
img_cv2 = cv2.imread(image_path)
print(img_cv2)
```

```python
plt.figure()
plt.imshow(img_cv2)
plt.axis('off')
plt.show()
```

```python
# reverse channels from BGR to RGB

#img_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
img_cv2_rgb = img_cv2[:, :, ::-1]

plt.figure()
plt.imshow(img_cv2_rgb)
plt.axis('off')
plt.show()
```

## 4. skimage

```python
from skimage import io

# skimage.io.imread loads the image, typically as an RGB array.
img_skimage = io.imread(image_path)
print(img_skimage)
```

```python
plt.figure()
plt.imshow(img_skimage)
plt.axis('off')
plt.show()
```
